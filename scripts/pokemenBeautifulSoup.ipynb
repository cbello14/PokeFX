```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd


url = "https://pokemondb.net/pokedex/all"
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
headers={'User-Agent':user_agent,} 

request=urllib.request.Request(url,None,headers) #The assembled request
page = urllib.request.urlopen(request)

html = page.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")
```

```python
pokemen = []
table = soup.find_all("table")[0]
rows = table.findChildren("tr", recursive=True)
rows.pop(0)
for row in rows:
    cells = row.findChildren("td")
    index = cells[0].get_text().strip()
    name = cells[1].get_text().replace(" ","_")
    name = name.replace("'","_")
    name = name.replace("%","_")
    name = ''.join(i for i in name if ord(i)<128)
    
    types = cells[2].get_text()
    link = "https://pokemondb.net"+cells[1].findChildren("a")[0]["href"]
    hp = cells[4].get_text()
    attack = cells[5].get_text()
    defense = cells[6].get_text()
    sp_atk = cells[7].get_text()
    sp_def = cells[8].get_text()
    spd = cells[9].get_text()
    pokemen.append({
        "name": name,
        "index":index,
        "types": types,
        "link": link,
        "moves": [],
        "stats": {"hp":hp, "atk":attack, "def":defense, "sp.atk":sp_atk, "sp.def":sp_def, "speed":spd}
    })

```

```python
pokemen
```

```python
for pokemon in pokemen:
    print(pokemon["name"], pokemen.index(pokemon))
    link = pokemon["link"]
    request=urllib.request.Request(link,None,headers) #The assembled request
    page = urllib.request.urlopen(request)

    html = page.read().decode("utf-8")
    soup = BeautifulSoup(html, "html.parser")
    tables = soup.find_all("table")
    for table in tables:
        if(table.get_text().count("Move") >0):
            rows = table.findChildren("tr", recursive=True)
            header = rows[0]
            headCells = header.get_text().split(" ")
            if ("Move" in headCells):
                index = headCells.index("Move")
            else:
                continue
            rows.pop(0)
            for row in rows:
                cells = row.findChildren("td")
                pokemon["moves"].append(cells[index].get_text())
    
    
```

```python
pokemen_df = pd.DataFrame(pokemen)
pokemen_df
```

```python
pokemen_df.to_csv("pokemen2.csv", sep=",", index=False, encoding="utf-8")
pokemen_alt = pd.read_csv("pokemen2.csv")
pokemen_alt.to_xml("pokemen2.xml", root_name="pokemen", index=False)
```
