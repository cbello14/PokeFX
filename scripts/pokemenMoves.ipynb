```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd


url = "https://pokemondb.net/move/all"
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
headers={'User-Agent':user_agent,} 

request=urllib.request.Request(url,None,headers) #The assembled request
page = urllib.request.urlopen(request)

html = page.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")
```

```python
moves = []
table = soup.find_all("table")[0]
rows = table.findChildren("tr", recursive=True)
rows.pop(0)
for row in rows:
    cells = row.findChildren("td")
    name = cells[0].get_text().replace(" ","_")
    name = name.replace("'","_")
    name = name.replace("%","_")
    name = ''.join(i for i in name if ord(i)<128)
    
    types = cells[1].get_text()
    link = "https://pokemondb.net"+cells[0].findChildren("a")[0]["href"]
    power = cells[3].get_text()
    acc = cells[4].get_text()
    pp = cells[5].get_text()
    moves.append({
        "name": name,
        "type": types,
        "link": link,
        "power": power,
        "accuracy" : acc,
        "pp": pp
    })
```

```python
moves
```

```python
moves_df = pd.DataFrame(moves)
```
